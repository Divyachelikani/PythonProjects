```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset using pandas
file_path = '/kaggle/input/upi-payment-transactions-dataset/transactions.csv'
d1 = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(d1.head())
```

```python
d1.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
sns.barplot(x='Timestamp',y='Amount (INR)',data=d1
plt.show()
```

```python
#group data by using hue
sns.barplot(x='Timestamp',y='Amount (INR)',data=d1,hue='Status')
plt.show()
```

```python
#palette
sns.barplot(x='Timestamp',y='Amount (INR)',data=d1,hue='Status',palette='rocket')
plt.show()
```

```python
#scatter plot

#palette
sns.scatterplot(x='Timestamp',y='Amount (INR)',data=d1,hue='Status')
plt.show()
```

```python
#distplot

#scatter plot

#palette
sns.distplot(d1['Amount (INR)'],color="green")
plt.show()
```

```python
#box plots
#scatter plot

#palette
sns.boxplot(x='Timestamp',y='Amount (INR)',data=d1, hue='Status', palette='rocket')
plt.show()
```
